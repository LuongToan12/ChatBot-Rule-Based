```python
import requests
```

```python
params = {
    "name" : "Toan",
    "age": 2
}
```

```python
#request.get
response = requests.get("https://httpbin.org/get", params = params)
print(response.url)

res_json = response.json()
del res_json['origin']
print(res_json)
```

```python
#request.post
payload = {
    "name" : "Toan",
    "age": 22
}

response = requests.post("https://httpbin.org/post", data = payload)
print(response.url)

res_json = response.json()
del res_json['origin']
print(res_json)
```

```python
#request.status
response = requests.post("https://httpbin.org/status/500")
if response.status_code == requests.codes.not_found:
    print("Not found")
else: 
    print(response.status_code)

#res_json = response.json()
#del res_json['origin']
#print(res_json)
```

```python
#User agent
header = {
    "User-Agent": "https://useragents.io/explore/platforms/ios/maker/apple-inc-b93/uas/mozilla-5-0-iphone-cpu-iphone-os-18-7-like-mac-os-x-applewebkit-605-1-15-khtml-like-gecko-version-26-2-1-mobile-15e148-safari-604-1-brave_b812add0b210316b97d97111a921e1d1",
    "Accept": "Image/jpeg"
}
response = requests.get("https://httpbin.org/image", headers=header)

with open("myimage.jpg","wb") as f:
    f.write(response.content)
```

```python
#User agent
for _ in [1,2,3]:
    try:
        response = requests.get("https://httpbin.org/delay/5", timeout=30)
    except:
        continue
res_json = response.json()
del res_json['origin']
print(res_json)
```
