```python
import nltk
import re
from nltk.chat.util import Chat
```

```python
reflections = {
    "tôi" : "bạn",
    "bạn" : "tôi",
    "mình": "bạn",
    "của tôi" : "của bạn",
    "của bạn" : "của tôi",
    "của mình" : "của bạn",
}
```
